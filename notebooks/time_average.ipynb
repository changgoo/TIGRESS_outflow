```python
tmin=200
tmax=500

mpdf_all={}
for phase in ['int','hot']:
    files = glob.glob('/tigress/changgoo/MHD_4pc_new/MVZ/{}/MHD_4pc_new_mvz.*.h5'.format(phase))
    files.sort()
    print(len(files))
    time=[]
    mpdf=[]
    for f in files:
        ds = yt.load(f)
        tMyr=ds.current_time.to('Myr').v
        if (tMyr >= tmin) & (tMyr <= tmax):
            print(tMyr)
            time.append(tMyr)
            mpdf.append(ds.data['cell_mass'].to('Msun').v)
        if tMyr > tmax: 
            break
    vbin=ds.data['x_bins'].v
    zbin=ds.data['y_bins'].v
    vc=0.5*(vbin[1:]+vbin[:-1])
    zc=0.5*(zbin[1:]+zbin[:-1])

    mpdf_xr=xr.DataArray(np.array(mpdf),coords=[time,vc,zc],dims=['time','vz','z'])

    mpdf_xr.to_netcdf('mpdf_{}.nc'.format(phase))
```

```python

def potential(z0):
    import astropy.constants as c
    import astropy.units as u

    G=c.G#.cgs.value
    pc=c.pc#.cgs.value
    Msun=c.M_sun#.cgs.value
    zstar=245.*pc
    rho_dm=6.4e-3 * Msun/pc**3
    R0=8.e3*pc
    surf_star=42*Msun/pc**2

    z   = z0 * pc
    phi = 2.*np.pi*G*surf_star*zstar*((1.+(z/zstar)**2.)**0.5 -1.) + 2*np.pi*G*rho_dm*R0*R0*np.log(1+(z/R0)**2.)
    return phi
```

```python
z=np.linspace(1000,3584,100)
phi=potential(z).to('km^2/s^2').value
```

```python
plt.plot(z,np.sqrt(2*(phi[-1]-phi)))
```
